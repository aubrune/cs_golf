```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gauss(x, mu, sigma):
    return np.exp(-(x-mu)*(x-mu)/(2*sigma*sigma))
```

```python
NUM_VALUES = 100
num_votes = 0
x = np.linspace(0, 1, NUM_VALUES)
scores = np.ones((NUM_VALUES, NUM_VALUES))
```

```python
plt.imshow(scores, cmap='hot', interpolation='nearest')
plt.show()
```

```python
def add_score(scores, mark, value_speed, value_angle, num_votes):
    SIGMA = 0.1
    speeds = np.tile((mark)*gauss(x, value_speed, SIGMA), [NUM_VALUES,1])
    angles = np.tile((mark)*gauss(x, value_angle, SIGMA).reshape((NUM_VALUES, 1)), NUM_VALUES)
    new_frame = np.multiply(speeds, angles)
    return (num_votes*scores + new_frame)/(num_votes+1), num_votes+1
```

```python
scores, num_votes = add_score(scores, 0.5, 0.4, 0.8, num_votes)
scores, num_votes = add_score(scores, 0.5, 0.8, 0.4, num_votes)
scores, num_votes = add_score(scores, 0.5, 0.8, 0.8, num_votes)
scores, num_votes = add_score(scores, 0.8, 0.4, 0.4, num_votes)
```

```python
plt.imshow(scores, cmap='hot', interpolation='nearest')
plt.show()
```
